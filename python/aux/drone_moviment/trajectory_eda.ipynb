```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df = pd.read_csv('./Positions.csv')
df = df.drop('foo', axis = 1)
df.head()
```

```python
fig = plt.figure(figsize=(20,15))

ax = fig.add_subplot(111, projection='3d')
pts = ax.scatter(df['x'], df['y'], df['z'], c = df['time'], cmap='OrRd', marker='x')

ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')

fig.colorbar(pts ,shrink=0.5, aspect=8)
plt.show()
```

```python
# 3D Grid space
grid_space = np.linspace(-2500,2500, 5000)
```

```python
class Grid3D:
    def __init__(self):
        self.space_1d = np.linspace(-25, 25, 5)
        self.grid = np.meshgrid(self.space_1d, self.space_1d, np.linspace(0, 50, 5))

grid3d = Grid3D()
```

```python
fig = plt.figure(figsize=(20,15))

ax = fig.add_subplot(111, projection='3d')
pts = ax.scatter(grid3d.grid[0], grid3d.grid[1], grid3d.grid[2], c='black', alpha=0.4, s =1)

ax.set_xlabel('x [m]')
ax.set_ylabel('y [m]')
ax.set_zlabel('z [m]')

# fig.colorbar(pts ,shrink=0.5, aspect=8)
plt.show()
```
